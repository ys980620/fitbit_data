```python
#changing date from string type to datetime tpye
```

```python
daily_activity['ActivityDate'] = pd.to_datetime(daily_activity['ActivityDate'])
hourly_calories['ActivityHour'] = pd.to_datetime(hourly_calories['ActivityHour'])
hourly_steps['ActivityHour'] = pd.to_datetime(hourly_steps['ActivityHour'])
```

```python
#creating column for day of the week and hour of day
```

```python
daily_activity['day_of_week'] = daily_activity['ActivityDate'].dt.day_name()
hourly_steps['hour'] = hourly_steps['ActivityHour'].dt.hour
hourly_calories['hour'] = hourly_calories['ActivityHour'].dt.hour
```

```python
#calculating average daily steps by day of the week
```

```python
avg_daily_steps = daily_activity.groupby('day_of_week')['TotalSteps'].mean().reset_index()
avg_daily_steps.sort_values(by = 'TotalSteps', ascending = False)
```

```python
#calculating average daily calories consumed by day of the week
```

```python
avg_daily_calories = daily_activity.groupby('day_of_week')['Calories'].mean().reset_index()
avg_daily_calories.sort_values(by = 'Calories', ascending = False)
```

```python
sns.scatterplot(x = daily_activity['ActivityDate'], y = steps['TotalSteps'])
```

```python
#calculating average daily steps and calories by day of the week from new dataframe without outliers
```

```python
avg_daily_steps = steps.groupby('day_of_week')['TotalSteps'].mean().reset_index()
avg_daily_steps.sort_values(by = 'TotalSteps', ascending = False)
```

```python
avg_daily_calories = calories.groupby('day_of_week')['Calories'].mean().reset_index()
avg_daily_calories.sort_values(by = 'Calories', ascending = False)
```

```python
steps_per_hr = hourly_steps.loc[hourly_steps['StepTotal'] <= 5000]
calories_per_hr = hourly_calories.loc[hourly_calories['Calories'] <= 500]
```

```python
#calculating average hourly steps and calories by hour of the day from new dataframe without outliers
```

```python
avg_steps_per_hr = steps_per_hr.groupby('hour')['StepTotal'].mean().reset_index()
avg_steps_per_hr.sort_values(by = 'StepTotal', ascending = False)
```

```python
avg_calories_per_hr = calories_per_hr.groupby('hour')['Calories'].mean().reset_index()
avg_calories_per_hr.sort_values(by = 'Calories', ascending = False)
```
